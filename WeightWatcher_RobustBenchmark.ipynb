```python
linf = ['Wong2020Fast','Engstrom2019Robustness','Salman2020Do_R50','Salman2020Do_R18', 'Standard_R50']
common_corruptions = ['Geirhos2018_SIN','Geirhos2018_SIN_IN','Geirhos2018_SIN_IN_IN','Hendrycks2020Many','Hendrycks2020AugMix','Standard_R50']
```

```python
model_linf =[None]*len(linf)
model_common_corruptions =[None]*len(common_corruptions)
i=0
for model_name in linf:
    model_linf[i] = load_model(model_name, dataset='imagenet', threat_model='Linf')
    i=i+1
```

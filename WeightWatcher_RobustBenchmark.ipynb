```python
linf = ['Wong2020Fast','Engstrom2019Robustness','Salman2020Do_R50', 'Standard_R50']
corruptions = ['Geirhos2018_SIN','Geirhos2018_SIN_IN','Geirhos2018_SIN_IN_IN','Hendrycks2020Many','Hendrycks2020AugMix','Standard_R50']
```

```python
Standard_Accuracy_Linf=[55.62, 62.56, 64.02, 76.52]
Rubust_Accuray_Linf=[26.24,29.22,34.96, 0.0]
Best_Known_Robust_Accuracy_Linf=[26.24,29.22,34.96, 0.0]
Standard_Accuracy_common_corruptions=[60.24,74.88,77.44,76.88,76.98,76.52]
Rubust_Accuray_In_C_common_corruptions=[39.53,45.52,41.72,52.65,48.31,39.23]
Rubust_Accuray_In_3DCC_common_corruptions=[37.80,48.09,46.92,54.13,51.35,44.77]
```

```python
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt
```

```python
metrics_dict = defaultdict(list)
for metric in ['alpha', 'entropy', 'mp_softrank', 'rand_distance', 'D', 'stable_rank']:

    for linf_name in linf:
        dataframe = pd.read_csv(f"ww_metrics/Linf/{linf_name}.csv")
        metrics_dict[metric].append(dataframe[metric].mean())

    plt.figure(figsize=(15, 3))
    plt.suptitle('Linf', y=1.1, fontsize=20)
    plt.subplot(1, 3, 1)
    plt.plot(range(len(linf)), Standard_Accuracy_Linf, marker='o')
    plt.title('Standard_Accuracy')
    plt.xticks(range(len(linf)), linf, rotation=45)
    plt.subplot(1, 3, 2)
    plt.plot(range(len(linf)), Rubust_Accuray_Linf, marker='o')
    plt.title('Rubust_Accuray')
    plt.xticks(range(len(linf)), linf, rotation=45)
    plt.subplot(1, 3, 3)
    plt.title(f'{metric}')
    plt.plot(range(len(linf)), metrics_dict[metric], marker='o')
    plt.xticks(range(len(linf)), linf, rotation=45)
    plt.plot()

```

```python
metrics_dict = defaultdict(list)
for metric in ['alpha', 'entropy', 'mp_softrank', 'rand_distance', 'D', 'stable_rank']:
    for linf_name in corruptions:
        dataframe = pd.read_csv(f"ww_metrics/corruptions/{linf_name}.csv")
        metrics_dict[metric].append(dataframe[metric].mean())

    plt.figure(figsize=(15, 3))
    plt.suptitle('Corruption', y=1.1, fontsize=20)
    plt.subplot(1, 4, 1)
    plt.plot(range(len(corruptions )), Standard_Accuracy_common_corruptions, marker='o')
    plt.title('Standard_Accuracy')
    plt.xticks(range(len(corruptions )), corruptions , rotation=45)
    plt.subplot(1, 4, 2)
    plt.plot(range(len(corruptions )), Rubust_Accuray_In_C_common_corruptions, marker='o')
    plt.title('Rubust_Accuray_In_C_common')
    plt.xticks(range(len(corruptions)), corruptions , rotation=45)
    plt.subplot(1, 4, 3)
    plt.plot(range(len(corruptions )), Rubust_Accuray_In_3DCC_common_corruptions, marker='o')
    plt.title('Rubust_Accuray_In_3DCC')
    plt.xticks(range(len(corruptions)), corruptions , rotation=45)
    plt.subplot(1, 4, 4)
    plt.title(f'{metric}')
    plt.plot(range(len(corruptions)), metrics_dict[metric], marker='o')
    plt.xticks(range(len(corruptions)), corruptions, rotation=45)
    plt.plot()

```
